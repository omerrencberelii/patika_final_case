```python
# we tried to concatenate all 5 data but we need to transform them to become similar
# we dropped all columns except country and happiness score
# we changed all column and row names in order to concatenate them
# at last, we concatenate all scores in horizontal direction
data2015=data2015.sort_values(by="Country").loc[:,["Country","Happiness Score"]].rename(columns={"Happiness Score":"Score2015"}).set_index('Country')
data2016=data2016.sort_values(by="Country").loc[:,["Country","Happiness Score"]].rename(columns={"Happiness Score":"Score2016"}).set_index('Country')
data2017=data2017.sort_values(by="Country").loc[:,["Country","Happiness.Score"]].rename(columns={"Happiness.Score":"Score2017"}).set_index('Country')
data2018=data2018.sort_values(by="Country or region").loc[:,["Country or region","Score"]].rename(columns={"Country or region":"Country", "Score":"Score2018"}).set_index('Country')
data2019=data2019.sort_values(by="Country or region").loc[:,["Country or region","Score"]].rename(columns={"Country or region":"Country", "Score":"Score2019"}).set_index('Country')
new_data=pd.concat([data2015,data2016,data2017,data2018, data2019 ],axis=1)
```

```python
#check the null values and drop them
new_data.isnull().sum()
```

```python
#after dropping null values, we need to redefine the data
new_data.dropna(inplace=True)
```

```python
#review the data, just first 5 row
new_data.head()
```

```python
#the difference in happiness score between 2015 and 2019
new_data["difference"] = (new_data["Score2019"]-new_data["Score2015"])
```

```python
#sorting the new_data from the highest difference to lowest difference
#indexing new_data again and create a "Country" column which includes indexes
#presenting barplot concerning descending difference in countries
#rotating x-parameters which is "Country" for -45 degree
#to sum up, this barplot show us 10 countries which vary from the highest happiness difference to the lowest happiness difference
new_data = new_data.sort_values("difference",ascending=False)
new_data["Country"] = new_data.index
ax = sns.barplot(data=new_data.head(10), x="Country", y="difference")
ax.tick_params(axis= "x", rotation= -45)
plt.title("Top 10 Country whose happiness is increasing most")
```

```python
#Similar algorithms but this time, we start presenting from the highest decrease in happiness to the lowest one
#In order to create nice plot, we converted minus happiness score to plus happiness score via "abs"
new_data=new_data.sort_values("difference")
new_data["difference"]=abs(new_data["difference"])
ax=sns.barplot(data=new_data.head(10),x="Country",y="difference")
ax.tick_params(axis = 'x', rotation = 45)
plt.title('Top 10 countries whose happiness is decreasing most')
```
